# TASK 5: Test of Benchmark of matrix multiplication

```python
import matplotlib.pyplot as plt
import numpy as np

files = ["matrix_64.txt", "matrix_128.txt", "matrix_256.txt", "matrix_512.txt"]
sequential_times = [0.04, 0.31, 2.56, 20.55]
map_reduce_times = [10.68, 47.50, 1410.17, 3172.88]
matrix_sizes = [64, 128, 256, 512]
```

```python
plt.style.use('seaborn-whitegrid')
colors = ['blue', 'green']

plt.figure(figsize=(12, 8))
plt.gca().patch.set_facecolor('white')

plt.plot(matrix_sizes, sequential_times, marker='o', label='Sequential', color=colors[0], linestyle='-', linewidth=2, markersize=10, alpha=0.8)
plt.plot(matrix_sizes, map_reduce_times, marker='o', label='MapReduce', color=colors[1], linestyle='-', linewidth=2, markersize=10, alpha=0.8)
plt.xticks(matrix_sizes, [str(size) for size in matrix_sizes])
plt.title('Sequential vs MapReduce Matrix Multiplication Times', loc='center', fontsize=18, fontweight=0, color='black')
plt.xlabel('Matrix Size', fontsize=14)
plt.ylabel('Time (ms)', fontsize=14)
plt.legend(fontsize=12)
plt.ylim(bottom=0)
plt.tight_layout()
plt.savefig('matrix_multiplication_times.svg', format='svg')
plt.show()
```
